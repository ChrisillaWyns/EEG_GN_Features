```python
#SGKF
import numpy as np
from sklearn.model_selection import StratifiedGroupKFold


def SGKF_split(X, y, subject_ids, n_splits=5):
    sgkf = StratifiedGroupKFold(
                n_splits=n_splits,
                shuffle=True,
                random_state=42)
    for fold, (train_idx, test_idx) in enumerate(
            sgkf.split(X,
                       y,
                       groups=subject_ids),1):
        print(f"\nFold {fold}")
        print("Training Subjects:",
              np.unique(subject_ids[train_idx]))

        print("Testing Subjects:",
              np.unique(subject_ids[test_idx]))
        yield train_idx, test_idx
```

```python
#LOSO
from sklearn.model_selection import LeaveOneGroupOut
def LOSO_split(X,
               y,
               subject_ids):
    logo = LeaveOneGroupOut()
    for subject,(train_idx,test_idx) in enumerate(

            logo.split(X,
                       y,
                       groups=subject_ids),1):

        print(f"\nLOSO Subject {subject}")

        print("Test Subject:",
              np.unique(subject_ids[test_idx]))
        yield train_idx,test_idx
```
