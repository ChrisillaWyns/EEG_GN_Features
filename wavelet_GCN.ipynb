```python
import numpy as np
import pywt

def wavelet_bandpass(data, wavelet='db2', level=7):
    """
    Wavelet-based 0.5–40 Hz filtering.
    data : ndarray
        Shape = (epochs, samples, channels)
    level :
        7 -> PhysioNet (160 Hz)
        8 -> BCI IV-2a / IV-2b (250 Hz)
    """

    epochs, samples, channels = data.shape

    filtered = np.zeros_like(data)

    for ep in range(epochs):

        for ch in range(channels):

            signal = data[ep, :, ch]

            coeffs = pywt.wavedec(
                signal,
                wavelet=wavelet,
                level=level,
                mode='periodization'
            )

        
            coeffs[0] = np.zeros_like(coeffs[0])

            
            coeffs[level] = np.zeros_like(coeffs[level])

            reconstructed = pywt.waverec(
                coeffs,
                wavelet=wavelet,
                mode='periodization'
            )

            filtered[ep, :, ch] = reconstructed[:samples]

    return filtered
```
